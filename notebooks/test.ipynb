```python
import requests

base_url = "http://127.0.0.1:8000"

# 1) login para obtener access token
login_resp = requests.post(
    f"{base_url}/api/v1/auth/login",
    json={"username": "jesus", "password": "jesus123"},
)
login_resp.raise_for_status()
access_token = login_resp.json()["access_token"]
```

```python
# 2) llamar endpoint protegido
report_resp = requests.post(
    f"{base_url}/api/v1/reports/generate_report",
    headers={"Authorization": f"Bearer {access_token}"},
    json={
        "chemical_compound": "C6H12O6",
        "prompt": "Analysis of chemical compound X",
        "title": "Chemical Analysis Report",
    },
)

print(report_resp.status_code)
print(report_resp.json())
```
